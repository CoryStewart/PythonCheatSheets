# COMMANDS

```python
import commands
help( commands.getstatusoutput )
```

```python
# Run command, wait for it to finish, returns tuple of: (status_int, {stderr + stdout combined})
(status, output) = commands.getstatusoutput( 'ls' )
print 'RESULT=', status
print 'OUTPUT=\n', output
```

# URLLIB

```python
import urllib
# Note: having env vars: http_proxy or https_proxy set when you're not behind a proxy will cause this to fail.
uf = urllib.urlopen( 'http://google.com/logos/doodles/2013/holiday-series-2013-2-5173241802915840-hp.png' )
# Download files with 'urlretrieve'
urllib.urlretrieve( 'http://google.com/logos/doodles/2013/holiday-series-2013-2-5173241802915840-hp.png', 'blah.png' )
# uf is kind of like a file handle that goes over the network
uf.read()
```

# Collections

```python
from collections import Counter   # this is one of the most useful modules

words = ['yes', 'but', 'no', 'but', 'yes']
wordcounts = Counter(words)
wordcounts['yes']
```

```python
wordcounts.most_common()    # counters do tabulation
```
